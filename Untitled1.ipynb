```python
import pandas as pd
import numpy as np

#Sample Dataset

data={'Name': ['Alice','Bob','Ava','Jmac'],
'Age': [25,np.nan, 30, np.nan],
'Salary': [50000,60000,None,45000]}

df=pd.DataFrame(data)
print("Before Cleaning:\n",df)
```

```python
#Drop Rows
df_drop = df.dropna()

#Fill
df['Age'].fillna(df['Age'].mean(),inplace=True)

#Fill Missing
df['Salary'].fillna(df['Salary'].median(),inplace=True)

print("\nAfter Cleaning:\n", df)
```

```python
import numpy as nm
import seaborn as sns
import matplotlib.pyplot as plt

#Example Data
data=[10,12,13,11,22,23,43,5,200,1900,86,92]

sns.boxplot(data)
plt.show()

#Detect using iqr
Q1,Q3=np.percentile(data,[25,75])
IQR= Q3-Q1


lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
outliers = [x for x in data if x < lower or x > upper]

print("Outliers Detected:",outliers)
```

```python
from sklearn .preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

df = pd.DataFrame({'Age': [18,22,40,32],
                   'Salary': [20000,50000,60000,80000]})

#Normalization
scaler_std = MinMaxScaler()
df_norm = scaler_std.fit_transform(df)

 #Standardizaton
scaler_std = StandardScaler()
df_std = scaler_std.fit_transform(df)


print("Original Data:\n",df)
print("\nNormalized Data:\n",df_norm)
print("\nStandardized Data:\n",df_std)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

#Simple Dataset
df = pd.DataFrame({
    'Name': ['Alice','Bob','Charles'],
    'Age': [25,30,40],
    'Country': ['USA','India','Israel'],
    'Spend1': [100,200,300],
    'Spend2': [150,250,450]

})

#Feature Engineering
df['Total Spend'] = df['Spend1'] + df['Spend2']

#Age
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 50], labels=['Young', 'Middle'])

#Encode
encoder = LabelEncoder()
df['Country_Encoded'] = encoder.fit_transform(df['Country'])

print(df)
```
